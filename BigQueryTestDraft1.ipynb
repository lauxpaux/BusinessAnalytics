```python
import pandas as pd
from google.cloud import bigquery
from google.colab import auth

auth.authenticate_user()

project_id = 'querytest2-444200'

# Construct a BigQuery client object.
client = bigquery.Client(project=project_id)

# Query the public dataset
query = """
    SELECT
      o.order_id,
      o.user_id,
      o.status,
      o.gender,
      o.created_at,
      o.num_of_item,
      e.*,
      oi.*,
      p.*,
      u.*
    FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.events` AS e
        ON o.user_id = e.user_id
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.order_items` AS oi
        ON o.order_id = oi.order_id
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.products` AS p
        ON oi.product_id = p.id
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
        ON o.user_id = u.id
    GROUP BY o.user_id, o.order_id
    LIMIT 10;
"""

# Run the query
query_job = client.query(query)

# Get the results
results = query_job.result()

# Convert to a Pandas DataFrame
df = results.to_dataframe()

df.head()
```

```python
import pandas as pd
from google.cloud import bigquery
from google.colab import auth

auth.authenticate_user()

project_id = 'querytest2-444200'

# Construct a BigQuery client object.
client = bigquery.Client(project=project_id)

# Query the public dataset
query = """
      WITH aggregated_events AS (
    SELECT user_id, COUNT(*) AS event_count, MAX(created_at) AS last_event_time
    FROM `bigquery-public-data.thelook_ecommerce.events`
    GROUP BY user_id
  ),
  aggregated_order_items AS (
    SELECT order_id, ARRAY_AGG(STRUCT(product_id, sale_price)) AS items
    FROM `bigquery-public-data.thelook_ecommerce.order_items`
    GROUP BY order_id
  )
  SELECT
      o.order_id,
      o.user_id,
      o.status,
      o.gender,
      o.created_at,
      ae.event_count,
      ae.last_event_time,
      ao.items,
      u.*
  FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
  LEFT JOIN aggregated_events AS ae
      ON o.user_id = ae.user_id
  LEFT JOIN aggregated_order_items AS ao
      ON o.order_id = ao.order_id
  LEFT JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
      ON o.user_id = u.id
  LIMIT 100;

"""

# Run the query
query_job = client.query(query)

# Get the results
results = query_job.result()

# Convert to a Pandas DataFrame
df2 = results.to_dataframe()

df2.head()
```

### Looping through full query

```python
import pandas as pd
from google.cloud import bigquery
from google.colab import auth


auth.authenticate_user()

fullDF = pd.DataFrame()

chunk_size = 10000
chunk = 0

for chunk in range(0, chunk_size):
  project_id = 'querytest2-444200'

  # Construct a BigQuery client object.
  client = bigquery.Client(project=project_id)

  # Query the public dataset
  query = """
    SELECT
      o.order_id,
      o.user_id,
      o.status,
      o.gender,
      o.created_at,
      o.num_of_item,
      e.*,
      oi.*,
      p.*,
      u.*
    FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.events` AS e
        ON o.user_id = e.user_id
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.order_items` AS oi
        ON o.order_id = oi.order_id
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.products` AS p
        ON oi.product_id = p.id
      LEFT OUTER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
        ON o.user_id = u.id
      LIMIT 1000;
  """

  # Run the query
  query_job = client.query(query)

  # Get the results
  results = query_job.result()

  # Convert to a Pandas DataFrame
  df = results.to_dataframe()

  # Appending dataframe
  fullDF = pd.concat([fullDF, df], ignore_index=True)

chunk+=1
fullDF.shape
```
