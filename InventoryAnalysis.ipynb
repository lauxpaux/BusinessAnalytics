 This means that if were to sell any of the items from the top ten biggest brands that make up our inventory, we'd have an average retail price of $55.27.

### Statistical Visualizations

Let's Investigate the distribution of values for Average Retail Price per Brand, Average Retail Price per Category.

```python
# selecting numerical columns
num_cols = products.select_dtypes(include='number')
num_cols.head()
```

Let's look at the distributions of values for the different columns.

#### Understanding Distributions for cost and retail price:

```python
cols = num_cols.columns

plt.figure(figsize=(10, 8))
plt.suptitle('Distribution of Numerical Columns', fontsize= '15')
#plt.subplots_adjust(top=0.7, wspace=1)

for col in range(len(cols)):
  plt.subplot(2, 2, col+1)
  sns.histplot(num_cols[cols[col]], kde=True, color='#0073e6')
  plt.title(cols[col])


  ax = plt.gca()
  #print(id(ax) == id(plt.gca()))

  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show();
```

We can ignore the `id` column for now and focus on the insights from the other distributions. Starting with `cost`: Off the bat, it looks like the majority of the items are about in the range of \$25-\$50 dollars. The distribution is right-skewed, with some very expensive items representing a long tail after ~\$120 dollars. We will investigate this more thoroughly shortly. The `retail_price` column shows a similar right skewed distribution with the cost of items ranging between $0 - \$100, and we have fewer items at a higher retail price. The multimodal distribution plot for distribution center indicates that this is a categorical variable with three distribution centers.

```python
fig = plt.figure(figsize=(10, 8))
plt.suptitle('Cost Distribution', fontsize= '15')

ax = plt.gca()
#print(plt.gca() == ax)

# plot
sns.histplot(products['cost'], kde=True, color='#0073e6', legend=True)
ax.axvline(x=products['cost'].mean(), color='red', linestyle='--')
ax.axvline(x=products['cost'].median(), color='blue', linestyle='-')

# customization
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.legend(labels=['Median', 'Mean'])
plt.show();

print(f'The mean cost is ${round(products["cost"].mean(), 3)}')
print(f'The median cost is ${round(products["cost"].median(), 3)}')
```

From the plot above, we can see that the median is a better representation of centrality, capturing the data points better.

The same is true about the median for retail price.

```python
fig = plt.figure(figsize=(10, 8))
plt.suptitle('Retail Price Distribution', fontsize= '15')

ax = plt.gca()
#print(plt.gca() == ax)

# plot
sns.histplot(products['retail_price'], kde=True, color='#0073e6', legend=True)
ax.axvline(x=products['retail_price'].mean(), color='red', linestyle='--')
ax.axvline(x=products['retail_price'].median(), color='blue', linestyle='-')

# customization
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.legend(labels=['Median', 'Mean'])
plt.show();


print(f'The mean cost is ${round(products["retail_price"].mean(), 3)}')
print(f'The median cost is ${round(products["retail_price"].median(), 3)}')
```

```python
boxplotdf = num_cols.drop(columns=['id', 'distribution_center_id'])
cols = boxplotdf.columns

plt.figure(figsize=(10, 8))
plt.suptitle('Box Plots', fontsize= '15')
#plt.subplots_adjust(top=0.7, wspace=1)

for col in range(len(cols)):
  plt.subplot(2, 2, col+1)
  sns.boxplot(boxplotdf[cols[col]], color='#0073e6')
  plt.title(cols[col])


  ax = plt.gca()
  #print(id(ax) == id(plt.gca()))

  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show();
```

As observed above, we continue to see similar trends for retail price and cost. If you recall, the median for cost is $19.115, and for retail price it was \$39.795, which coincide with the boxplots we can see. From this plot we observed a lot of outliers.
