Because the user ids don't match when we perform the data call. We'll need to match the user ids for both table when the data is being pulled.

```python
auth.authenticate_user()

project_id = 'querytest2-444200'

# Construct a BigQuery client object.
client = bigquery.Client(project=project_id)

# Query the public dataset
query = """

  SELECT
      u.*, o.*
  FROM
      `bigquery-public-data.thelook_ecommerce.users` AS u
  JOIN
      `bigquery-public-data.thelook_ecommerce.orders` AS o
  ON
      u.id = o.user_id
  WHERE
      u.id IS NOT NULL AND o.user_id IS NOT NULL

"""

# Run the query
query_job = client.query(query)

# Get the results
results = query_job.result()

# Convert to a Pandas DataFrame
JoinTest = results.to_dataframe()

JoinTest.head()
```
